Feature engineering and extraction

```python
df_clean['bmi_catergory'] = pd.cut(
    df_clean['bmi'],
    bins = [0,18.5,24.9,29.9,float('inf')],
    labels = ['Underweight','Normal','Overweight','obese']
)
```

```python
df_clean
```

```python
df_clean.rename(columns={'bmi_catergory': 'bmi_category'}, inplace=True)
```

```python
df_clean = pd.get_dummies(df_clean,columns=['bmi_category'])
```

```python
df_clean = df_clean.astype(int)
df_clean.head()
```

```python
import sys
!{sys.executable} -m pip install scikit-learn
```

```python
from sklearn.preprocessing import StandardScaler
cols = ['age','bmi','children']
scaler = StandardScaler()

df_clean[cols] = scaler.fit_transform(df_clean[cols])
```

```python
df_clean
```

```python
import scipy
from scipy.stats import pearsonr

selected_features = [
    'age','bmi','children','sex','smoker','region_northwest','region_southwest','region_northeast','region_southeast',
    'bmi_category_Normal','bmi_category_Underweight','bmi_category_Overweight','bmi_category_obese']

correlations = {
    feature: pearsonr(df_clean[feature], df_clean['charges'])[0]
    for feature in selected_features
}

correlation_df = pd.DataFrame(list(correlations.items()),columns=['Features','Pearson Correlation'])
correlation_df.sort_values(by='Pearson Correlation',ascending=False)

```

```python
cat_features = [
    'sex','smoker','region_northwest','region_southwest','region_southeast',
    'bmi_category_Normal','bmi_category_Overweight','bmi_category_obese'
]
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

alpha = 0.05
df_clean['charges_bin'] = pd.qcut(df_clean['charges'],q=4,labels=False)
chi2_results = {}

for col in cat_features:
    contingency = pd.crosstab(df_clean[col], df_clean['charges_bin'])
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')
chi2_df
```

```python
final_df = df_clean[['age', 'sex', 'bmi', 'children', 'smoker', 'charges','region_southeast','bmi_category_obese']]
final_df
```
